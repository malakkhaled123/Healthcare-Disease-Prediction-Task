# Cleaning & Merging Lab tests Datasets

```python
import pandas as pd
```

```python
# Glycohemoglobin
df_hba1c = pd.read_sas('/content/drive/MyDrive/Healthcare Prediction Task/data/raw/lab tests/GHB_L.xpt')

# Plasma Fasting Glucose
df_glucose = pd.read_sas('/content/drive/MyDrive/Healthcare Prediction Task/data/raw/lab tests/GLU_L.xpt')

# Insulin
df_insulin = pd.read_sas('/content/drive/MyDrive/Healthcare Prediction Task/data/raw/lab tests/INS_L.xpt')

```

```python
#Extracting Relevant columns from each dataset for merging

df_hba1c = df_hba1c[['SEQN', 'LBXGH']].rename(columns={'SEQN':'ID','LBXGH': 'HbA1c'})
df_glucose = df_glucose[['SEQN', 'LBXGLU']].rename(columns={'SEQN':'ID','LBXGLU': 'Glucose'})
#df_insulin = df_insulin[['SEQN', 'LBXIN']].rename(columns={'SEQN':'ID','LBXIN': 'Insulin'})
```

```python
# Merging Datasets into one Lab tests dataset
dfs = [df_hba1c, df_glucose]

# First, set 'ID' as index for each dataframe
for df in dfs:
    df.set_index('ID', inplace=True)

# Concatenate along columns
df_lab = pd.concat(dfs, axis=1)
df_lab.reset_index(inplace=True)
```
