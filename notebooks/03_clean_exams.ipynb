```python
#Extracting Relevant columns from Body Measures dataset for merging
df_BodyMeasures = df_BodyMeasures[['SEQN','BMXBMI','BMXWT','BMXHT' ,'BMXWAIST']].rename(columns = {
    'SEQN' : 'ID',
    'BMXBMI' : 'BMI',
    'BMXWT' : 'weight',
    'BMXHT' : 'height',
    'BMXWAIST' : 'waist'
})
```

```python
#Extracting Relevant columns from Blood Pressure dataset for merging
df_BloodPressure = df_BloodPressure[['SEQN','BPXOSY1', 'BPXOSY2', 'BPXOSY3','BPXODI1', 'BPXODI2', 'BPXODI3']]

# Extracting systolic BP , diastolic BP as an average of their 3 readings
df_BloodPressure['systolic_bp'] = df_BloodPressure[['BPXOSY1','BPXOSY2','BPXOSY3']].mean(axis=1)
df_BloodPressure['diastolic_bp'] = df_BloodPressure[['BPXODI1','BPXODI2','BPXODI3']].mean(axis=1)


```

```python
# drop raw readings
df_BloodPressure = df_BloodPressure[['SEQN','systolic_bp','diastolic_bp']].rename(columns = {'SEQN' : 'ID'})
```

```python
# Merging Datasets into one examinations dataset
dfs = [df_BodyMeasures, df_BloodPressure]

# First, set 'ID' as index for each dataframe
for df in dfs:
    df.set_index('ID', inplace=True)

# Concatenate along columns
df_exam = pd.concat(dfs, axis=1)
df_exam.reset_index(inplace=True)
```
